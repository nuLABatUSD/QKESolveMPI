```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
organized_data = {}
rows = []

with open('R_values_test.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile)
    
    for i,row in enumerate(csvreader):
        if i%2==0:
            my_eps_value = []
            eps = row[2:]
            my_key = (float(row[0]), float(row[1]))
            for e in eps:
                my_eps_value.append(float(e))
            
        else:
            my_dens_value = []
            for e in range(0, len(row)-1):
                my_dens_value.append(float(row[e]))
            organized_data[my_key] = [my_eps_value, my_dens_value]
```

```python
plt.figure()
for key in organized_data:
    if key ==(20,200):
        plt.plot(organized_data[key][0][1:], organized_data[key][1][1:], label=key)
plt.legend()
plt.xlabel("Epsilon")
plt.ylabel("R")
plt.show()
```
